Implement the Czekanowski-Dice index. 

\begin{equation}
C=2\frac{P\times R}{R+R}.
\end{equation}
Precision is defined as:
\begin{equation}
PPV=\frac{\#TP}{\#TP+\#FP}.
\end{equation}
Recall is defined as:
\begin{equation}
TPR=\frac{\#TP}{\#TP+\#FN}.
\end{equation}


```python

import numpy
dataset1 = numpy.random.randint(0, 2, 50)#['n','i','g','h','t']#
dataset2 = numpy.random.randint(0, 2, 50)#['n','a','c','h','t']#numpy.random.randint(0, 4, 20)
#print(dataset1)
#print(dataset2)
    
def dice3( original, second):
    
    comon = []
    
    for i in range(len(original)):
        if original[i] == second[i]:
            comon.append(original[i])
    
    a = len(comon)
    b = len(second)
    c = len(original)
    
    d_indx = 2*a/(b+c)
    
    return d_indx
    
    
def dice_score(  original,second ):
    
    A = 0
    BminusA = 0
    AcommonB = 0
    
    if( len(original) != len(second)): 
        return 0;
    
    for i in range(len(original)):
        if original[i] == 1:
            A += 1
        if original[i] == 1 and second[i] == 1:
            AcommonB += 1
        if original[i] == 0 and second[i] == 1:
            BminusA  +=1
            
    #print( AcommonB, A, BminusA)
    
    d_indx = AcommonB/(A + BminusA)
    
    return d_indx

def dice_ind( original, second):
  
    #Almost for sure incorect version

    bigramsO = []
    bigramsS = []
    
    for i in range(len(original)-1):
        if not [original[i], original[i+1]] in bigramsO:
            bigramsO.append( [original[i], original[i+1]])
        
    for i in range(len(second)-1):
        if not [second[i], second[i+1]] in bigramsS:
            bigramsS.append( [second[i], second[i+1]])
        
    matched = 0.0
    
    #print(bigramsS)
    
    #print(bigramsO)
    
    for i in bigramsS:
        if i in bigramsO:
            matched += 1
        
    d_indx = 2*matched/(len(bigramsO) + len(bigramsS))
    
    return d_indx

#print(dice_ind(dataset1,dataset2))
#print(dice_score(dataset1,dataset2))





from sklearn.metrics import f1_score
print(f1_score(dataset1, dataset2))

from sklearn.metrics import jaccard_similarity_score
print("Sklearn :", jaccard_similarity_score(dataset1, dataset2))


print("Mine :",dice3(dataset1,dataset2))
```
